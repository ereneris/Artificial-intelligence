```python
import tkinter as tk
from tkinter import filedialog
from tkinter import messagebox
import numpy as np
from PIL import Image, ImageDraw
import cv2
import tensorflow as tf
from keras.models import load_model
```

```python

loaded_model = load_model('my_model.h5')
```

```python
def save_image():
    global tk_image
    tk_image.save('kk.png')
    original_image = cv2.imread("kk.png")
    # Resmi siyah beyaza dönüştürme
    grayscale_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2GRAY)

    # Resmi 28x28 boyutunda yeniden boyutlandırma
    smaller_image = cv2.resize(grayscale_image, (28, 28))
    print(smaller_image)
    # CNN modeli 3D bir tensör bekler, bu nedenle boyut eklememiz gerekiyor
    input_image = np.expand_dims(smaller_image, axis=0)
    input_image = np.expand_dims(input_image, axis=3)

    # Tahmin yapma
    probability = loaded_model.predict(input_image)
    predicted_class = np.argmax(probability)

    # Sonucu gösterme
    result = messagebox.showinfo("Sonuç", "Tahmin Edilen Sınıf: {}".format(predicted_class))
```

```python
def paint(event):
    x1, y1 = (event.x - 1), (event.y - 1)
    x2, y2 = (event.x + 1), (event.y + 1)
    canvas.create_oval(x1, y1, x2, y2, fill="white", width=12)  # Çizim rengi beyaz olacak
    draw.line([x1, y1, x2, y2], fill="white", width=12)  # Çizim rengi beyaz olacak
```

```python
def clear_canvas():
    canvas.delete("all")
    tk_image = Image.new("L", (280, 280), "black")  # Arka plan rengini beyaz yap
    global draw
    draw = ImageDraw.Draw(tk_image)
```

```python

root = tk.Tk()
root.title("Çizim Formu")

# Arka plan rengini siyah yap
root.configure(bg='white')

canvas = tk.Canvas(root, width=280, height=280, bg='white')
canvas.pack()

tk_image = Image.new("L", (280, 280), "black")  # Arka plan rengini beyaz yap
draw = ImageDraw.Draw(tk_image)

canvas.bind("<B1-Motion>", paint)

clear_button = tk.Button(root, text="Temizle", command=clear_canvas)
clear_button.pack()

save_button = tk.Button(root, text="Tahmin Et", command=save_image)
save_button.pack()

root.mainloop()
```
