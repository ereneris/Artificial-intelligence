## Visualtion Engine

```python
df.groupby("Origin Station")["Trip Count"].value_counts()
```

```python
# Hafta günlerine Göre Toplam Yolcu Sayısının Grafiksel Dağılımı
day_trip_counts = df.groupby("Day_name")["Trip Count"].sum().reset_index()

sns.barplot(x='Day_name', y='Trip Count', data=day_trip_counts)
plt.show()
#En yoğun yolcu girişi Salı Çarşambe Ve Perşembe günleridir
```

```python
# Yıllara Göre Yolcu Sayısı Değişimi
total_count = df.groupby("Year")["Trip Count"].sum().reset_index()
sns.lineplot(x='Year', y='Trip Count', data=total_count)
plt.show()
#2020 Yılında pandemiyle birlikte toplu taşımayı kullanların sayısında azalma mevcut
```

```python
# İstasyonlara Göre Yolcu Sayısı Değişimi
plt.figure(figsize=(10, 6))
total_origin_count = df.groupby("Origin Station")["Trip Count"].sum().reset_index()
sns.barplot(x='Origin Station', y='Trip Count', data=total_origin_count)
plt.xticks(rotation=90)
plt.show()
#En çok yolcu girişi yapılan istasyon grafiktende anlaşılacağı üzere EMBR İstasyonudur.
```

```python
selected_hours_df = df[(df['Hour'] >= 22) | (df['Hour'] < 6)]
total_hour_count = selected_hours_df["Trip Count"].sum()
print("Istasyonda En gec Saatlerde Binen yolcu Kişi Sayısı:",total_hour_count)
```
